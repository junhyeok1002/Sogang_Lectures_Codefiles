```python
import pickle
import pandas as pd, numpy as np

results = list()

# load
for i in range(1,11):
    with open(f'data_list_seed{i}.pickle', 'rb') as f:
        temp = pickle.load(f)
        results.append(temp)
        
total = list()
for k in range(len(results[0])):
    sum = 0
    for i in range(len(results)):
        sum += results[i][k]
    mean = sum/len(results)
    total.append(mean)
total
```

### TEST3 : Making Missing Value
- 목적 : sparse matrix까지는 아니지만 결측치를 대거 생성하여 결측치가 많아졌을때 모형의 성능들이 어떻게 변하는 지 확인하기 위함
<br><br>

< 관찰 포인트 > 
- XGBoost와 LightGBM이 결측치가 많은 데이터에 강할 것이라는 예측과는 다르게 세 모형 전부 비슷한 수준으로 성능이 저하되었다.
- 그러나 특징적인 부분은 성능의 저하정도가 결측치 처리를 할 수 있는 XGB와 LGB가 비슷한 수준으로 저하되었고, 결측치 처리하지 못하는 CAT은 두 모형보다 조금 더 저하되었다
- 제대로된 효과를 확인하려면 하이퍼파라미터 튜닝으로 모형의 성능을 더 끌어 올리거나 진짜 sparse matrix자료에 이용해 확인하여야 할 것같다.
- 또한 Catboost가 튜닝속도도 결측치가 있는 경우에 더 빨랐다. 이것은 앞의 두 모형이 loss를 이용해 결측치를 메워나가는 방식때문에 속도가 느려진 것으로 보인다. XGB와 LGB에게 결측치는 처리해야할 대상이지만 CAT에게는 학습할 데이터가 감소하는 느낌이기 때문이다.
