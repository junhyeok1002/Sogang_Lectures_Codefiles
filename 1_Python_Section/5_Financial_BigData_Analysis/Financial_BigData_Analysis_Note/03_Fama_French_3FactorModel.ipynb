# Fama-French 3-Factor Model(1992) Empirics

\begin{align*}
(r_i - r_f) = \alpha + \beta (r_M - r_f) + \theta_s SMB + \theta_v HML +\epsilon
\end{align*}

[FAMA-French Library](http://mba.tuck.dartmouth.edu/pages/faculty/ken.french/data_library.html)

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

# factors(market, smb, hml)
factors = pd.read_csv('F-F_Research_Data_Factors.csv',index_col=0, header=3).dropna()
factors.columns = ['market', 'smb', 'hml', 'rf'] # macket 수익률 , smb, hml, risk free 수익률
factors.head()
```

```python
factors.index = factors.index.map(lambda x: datetime.strptime(str(x), "%Y%m")) #위의 숫자(192607)로된 index를 변경
factors.head()
```

```python
# 6 portfolios (size: 2, value: 3)

ptf = pd.read_csv('6_Portfolios_2x3.csv',index_col=0, header=15).dropna()
ptf.index = ptf.index.map(lambda x: datetime.strptime(str(x), "%Y%m"))
ptf.columns = ['s_bm1','s_bm2','s_bm3','b_bm1','b_bm2','b_bm3'] #월별 수익률임!
#s_bm => small시가총액, 뒤의 숫자 123은 베타크기에 따른 분류(3이제일 큼)
ptf.head()
```

```python
sns.pairplot(factors) #신경쓰지 않아도 됨
```

```python
# performances of potfolios

plt.figure(figsize=(12,8))
for col in ptf.columns :
    plt.plot(ptf[col].cumsum(), label=col) #누적수익률
plt.legend()
plt.show()
```

```python
df = pd.concat([ptf, factors], axis=1)
df.head()
```

```python
import statsmodels.api as sm


y = df['s_bm1'] 
X = df[['market', 'smb', 'hml']]

ols = sm.OLS(y,sm.add_constant(X)).fit()
ols.summary()
```
