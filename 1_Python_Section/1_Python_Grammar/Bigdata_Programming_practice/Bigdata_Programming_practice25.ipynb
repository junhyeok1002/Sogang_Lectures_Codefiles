```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

with plt.xkcd():
    plt.title('XKCD style plot!!!')
    X = np.linspace(-np.pi, np.pi, 1000)
    C = np.cos(X)
    plt.plot(X,C)
    plt.xticks(np.linspace(-3,3,7, dtype =np.int32))
    plt.yticks(np.linspace(-1,1,9))
    plt.annotate('-0.5 Here!',xy =(np.pi*2/3,-0.5),xycoords ='data', xytext=(0.2,-1),\
                fontsize = 16, arrowprops=dict(arrowstyle="->"))
    plt.scatter([np.pi*2/3],[-0.5], color ='r', linewidth = 3)
plt.show()
```
