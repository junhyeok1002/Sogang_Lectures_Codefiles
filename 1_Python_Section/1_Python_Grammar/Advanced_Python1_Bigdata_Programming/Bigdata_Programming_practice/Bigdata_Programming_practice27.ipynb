```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

Lee = np.array([75,89,92,78])
Park = np.array([90,76,88,83])
Jung = np.array([55,99,81,88])
Choi = np.array([80,67,75,91])
stu = np.array([Lee,Park,Jung,Choi])
subjects = np.array(["국어","수학","영어","컴"])

subject_avg = (Lee+Park+Jung+Choi)/4
bests = np.max(stu,axis=0)
totals = np.sum(stu, axis = 1)
```

```python
print("총점 반 평균:",np.sum(totals)/4)
print("\n각 과목의 평균과 최고점")
for i in range(len(subjects)):
    print("\t{}: {:.2f}, {}".format(subjects[i],subject_avg[i],bests[i]))
```

```python
with plt.xkcd():
    plt.title('Average score of Students')
    plt.ylabel('Average score') ;plt.xlabel('Students')
    Y = totals/4 ; X = ["Lee","Park","Jung","Choi"]
    plt.scatter(X,Y, color ='b', linewidth = 4)
    plt.plot(X,Y)
plt.show()
```

```python
with plt.xkcd():
    plt.title('Average and Top score of Subjects')
    plt.ylabel('Score'); plt.xlabel('Subjects')
    Y1 = subject_avg ; Y2 = bests ; X = ["Kor","Math","Eng","Com"]
    plt.plot(X,Y1,label = "subject avg"); plt.plot(X,Y2,label = "subject max")
    plt.legend(loc=0)
plt.show()
```
