# 4주차 실습8

```python
import csv
import matplotlib.pyplot as plt
import math
import matplotlib
import numpy as np
matplotlib.rcParams['axes.unicode_minus'] = False

dong = "제주특별자치도"

f = open("./gender.csv")
data = csv.reader(f)
m = []
f = []

for row in data:
    if len(m) == 101: break
    if dong in row[0]:
        for i in range(0,101):
            temp_m = row[i+3].replace(",","")
            temp_f = row[-(i+1)].replace(",","")
            m.append(int(temp_m))
            f.append(int(temp_f))
f.reverse()

plt.figure(figsize =(12,6))
plt.rc('font', family='Malgun Gothic')
matplotlib.rcParams['axes.unicode_minus'] = False
plt.scatter(m,f, c= range(101),s = [math.sqrt(m[i]+f[i]) for i in range(len(m))], alpha = 0.7, cmap = 'jet')
plt.plot(range(max(m)), range(max(m)), 'g')
plt.colorbar()
plt.xlabel("남성 인구 수") ; plt.ylabel("여성 인구 수")
plt.title("제주 지역의 성별 인구 그래프")
# plt.title(f"{dong} 지역의 남녀 성별 비율")
plt.style.use('ggplot')
plt.show()
```

## 5주차 실습2

```python
f =open('./age_no_comma.csv')
data = csv.reader(f)
data = list(data)

D_dict = dict()
for d in data[1:]:
    d_list = list()
    for i in d[3:10]:
        d_list.append(int(i.replace(",","")))
    if int(d[1].replace(",","")) == 0 : continue
    D_dict[d[0]] = sum(d_list)/(int(d[1].replace(",","")))
best = (sorted(D_dict.items(), key=lambda x: x[1], reverse=True)[0])
best
```

```python
f =open('./age_no_comma.csv')
data = csv.reader(f)
next(data)
name = best[0]
for row in data :
    if name in row[0] :
        home = np.array(row[3:], dtype = int) # 데이터 타입을 정수로
        
plt.style.use('ggplot')
plt.figure(figsize = (10,5))
plt.rc('font', family ='Malgun Gothic')
plt.title(name +' 지역의 인구 구조')
plt.plot(home)
print(f"{best[0]}지역의 0세 이상 6세 이하의 인구비율 : {round(best[1]*100,1)}")
plt.show()
```
