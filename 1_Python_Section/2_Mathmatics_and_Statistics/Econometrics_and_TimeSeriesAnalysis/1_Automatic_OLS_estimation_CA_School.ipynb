# OLS estimation

This program estimates ARX models for GDP, consumption, investment and export growth rates.

##  Import necessary packages

```python
import numpy as np
import statsmodels.api as sm
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

##  Read data

```python
from pandas import read_excel
data = read_excel('caschool.xlsx','caschool')
data
```

```python
data['testscr']=np.log(data['testscr']) 
data['str']=np.log(data['str'])
data['avginc']=np.log(data['avginc'])
```

## OLS estimation

```python
index_set=[['str'], ['avginc'], ['str', 'avginc']]
for i in index_set:
  model=sm.OLS(data['testscr'],sm.add_constant(data[i]))
  results=model.fit()
  print(results.summary())
```
