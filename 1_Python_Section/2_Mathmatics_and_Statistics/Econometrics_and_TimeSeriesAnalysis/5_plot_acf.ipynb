```python
phi=0.3
ts_length=300
current_x=0
x_values=[]
for i in range(ts_length+1):
    x_values.append(current_x)
    current_x=phi*current_x + np.random.randn()
```

```python
plt.plot(range(ts_length+1),x_values)
plt.show()
```

```python
n_lags=40
atcf_qs=stattools.acf(x_values, nlags=n_lags,qstat=True)
  # atcf_qs[0]: values of ACF
  # atcf_qs[1]: values of q-stat
  # atcf_qs[2]: p-values of the q-stat
temp=n_lags+1
plt.bar(range(temp),atcf_qs[0])
plt.show()
```
