# Trends of  GDP 

## 1. Import necessary packages

```python
import numpy as np
import statsmodels.tsa.arima_model as smta
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from pandas import read_excel
import seaborn as sns
import statsmodels.api as sm
```

## 2. Read data

```python
excel_data= read_excel('국내총생산에_대한_지출_원계열__실질__연간__20190408.xlsx', '데이터')
# excel_data
```

## 3. Extract level data

```python
# The type of these data is Series.
gdpl=excel_data.iloc[-1,1:]
```

## 4. Calculate percentage change

```python
gdp=gdpl.pct_change(periods=1)
gdp=100*gdp
```

## 5. Plot data

```python

dates = pd.date_range('1961', periods=58, freq='A')
fig=plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(dates,gdp[1:], 'k', label='GDP')
# ax.plot(dates,invst[1:], 'k--', label='Investment')
# ax.legend( loc =' best')
ax.set_title('GDP and investment Growth Rates (1961-2018)')
plt.show()
```

## 6. Trend estimation for GDP growth rates

```python
# 1961-2018
y=gdp[1:]
ss=len(y)
tr = np.arange(1,ss+1)
trs= tr*tr
tt = np.vstack((tr,trs))
model=sm.OLS(y,sm.add_constant(tt.T))
results=model.fit()
print(results.summary())
```

```python
# 1970-2018
y=gdp[10:]
ss=len(y)
tr = np.arange(1,ss+1)
trs= tr*tr
tt = np.vstack((tr,trs))
model=sm.OLS(y,sm.add_constant(tt.T))
results=model.fit()
print(results.summary())
```
